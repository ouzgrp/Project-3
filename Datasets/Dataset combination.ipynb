```python
sums = {}
with open("country-opendata-2016-2017A.csv", "r" ) as fp :
    for line in fp.readlines()[1:]:
        token = line.split(",")
        country_token = token[1].strip()

        if " - " in line:
            amount = 0.0
        elif '"' in line:
            amount = float(line.split('"')[1].strip().replace(",", ""))
        else:
            amount = float(line.split(",")[3].strip())

        if country_token not in sums:
            sums[country_token] = 0.0
        sums[country_token] += amount
```

```python
document = ""
with open("combinedData2.csv", "r") as fp:
    line_number = -1
    for line in fp.readlines():
        line_number += 1
        if line_number == 0:
            document += line + "\n"
            continue

        country_token = line.split(",")[-3].strip()
        if country_token in sums:
            amount = sums[country_token]
            line_changed = line.replace(line.split(",")[-1], str(amount))
            document += line_changed + "\n"

with open("canada_aid_combined.csv", "w") as fp:
    fp.write(document)
```
